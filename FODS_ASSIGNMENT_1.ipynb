```python
# -*- coding: utf-8 -*-
"""FODS_ASSIGNMENT_finale.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1PXM4bulBTY5P1mnyvtCwWPDryqW0nCL5

IMPORTING IMPORTANT LIBRARIES
"""

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import random

"""MOUNTING DRIVE"""

from google.colab import drive
drive.mount("/content/gdrive")

"""FUNCTION DEFINITIONS"""

def cost_function(X,y,w,lam):
  hypothesis = np.dot(X,w.T)
  J = (1/(2*len(y))) * np.sum((hypothesis - y)**2)
  return J

def cost_function_L2(X,y,w,lam):
  hypothesis = np.dot(X,w.T)
  J = (1/(2*len(y))) * np.sum((hypothesis - y)**2) + lam*np.sum(w**2)/2
  return J

def cost_function_L1(X,y,w,lam):
  hypothesis = np.dot(X,w.T)
  J = (1/(2*len(y))) * np.sum((hypothesis - y)**2) + lam*np.abs(np.sum(w))/2
  return J

def batch_gradient_descent(X,y,w,alpha,iters):
  cost_history = np.zeros(iters)
  for i in range(iters):
    hypothesis = np.dot(X,w.T)
    w = w - (alpha/len(y)) * np.dot(hypothesis - y,X)
    cost_history[i] = cost_function(X,y,w,lam)
  return w,cost_history

def batch_gradient_descent_L2(X,y,w,alpha,iters,lam):
  cost_history = np.zeros(iters)
  for i in range(iters):
    hypothesis = np.dot(X,w.T)
    w = (1-alpha*lam)*w - (alpha/len(y)) * np.dot(hypothesis - y,X)
    cost_history[i] = cost_function_L2(X,y,w,lam)
  return w,cost_history

def batch_gradient_descent_L1(X,y,w,alpha,iters,lam):
  cost_history = np.zeros(iters)
  for i in range(iters):
    hypothesis = np.dot(X,w.T)
    w = (1-(np.sign(np.sum(w)))*alpha*lam/2)*w - (alpha/len(y)) * np.dot(hypothesis - y,X)
    cost_history[i] = cost_function_L1(X,y,w,lam)
  return w,cost_history

def stochastic_gradient_descent(X,y,w,alpha,iters):
  cost_history = np.zeros(iters)
  for i in range(iters):
    rand_index = np.random.randint(len(y))
    ind_x = X[rand_index:rand_index + 1]
    ind_y = y[rand_index:rand_index + 1]
    w = w - alpha * (ind_x.T.dot(ind_x.dot(w) - ind_y))
    cost_history[i] = cost_function(ind_x,ind_y,w,lam)
  return w,cost_history

def stochastic_gradient_descent_L2(X,y,w,alpha,iters,lam):
  cost_history = np.zeros(iters)
  for i in range(iters):
    rand_index = np.random.randint(len(y))
    ind_x = X[rand_index:rand_index + 1]
    ind_y = y[rand_index:rand_index + 1]
    w = (1-alpha*lam)*w - alpha * (ind_x.T.dot(ind_x.dot(w) - ind_y))
    cost_history[i] = cost_function_L2(ind_x,ind_y,w,lam)
  return w,cost_history

def stochastic_gradient_descent_L1(X,y,w,alpha,iters,lam):
  cost_history = np.zeros(iters)
  for i in range(iters):
    rand_index = np.random.randint(len(y))
    ind_x = X[rand_index:rand_index + 1]
    ind_y = y[rand_index:rand_index + 1]
    w = (1-(np.sign(np.sum(w)))*alpha*lam/2)*w - alpha * (ind_x.T.dot(ind_x.dot(w) - ind_y))
    cost_history[i] = cost_function_L1(ind_x,ind_y,w,lam)
  return w,cost_history

def Normalise(X):
  mean = X.mean()
  std = np.std(X,axis=0)
  m = len(X)
  X_norm = (X-mean)/std
  return X_norm

def create_poly(a, b, degree, X_norm):
  
  count = 0
  for i in range(degree+1):
    for j in range(degree-i+1):
      count = count + 1
      X_norm = np.c_[X_norm,(a**i)*(b**j)]
  w = np.zeros((count))

  return X_norm,w

def Test(X_test,i):
    X_test_norm = Normalise(X_test)
    X_test_norm = X_test_norm.to_numpy()
    X_test_norm,ab = create_poly(X_test_norm[:,0],X_test_norm[:,1],i,X_test_norm)
    X_test_norm = X_test_norm[:,2:]
    return X_test_norm

def print_metrics(X_test_norm,batch_w,y_test):
    print('mean squared error :',mean_squared_error(np.dot(X_test_norm,batch_w.T),y_test))
    print('mean absolute error :',mean_absolute_error(np.dot(X_test_norm,batch_w.T),y_test))
    print('correlation coefficient :',correlation_coefficient(np.dot(X_test_norm,batch_w.T),y_test)) 
    print('--------------------------------------------------------\n')

def plot(title,iters,J_his_batch_S,i):
    plt.plot(range(iters),J_his_batch_S)
    plt.title(title)
    plt.xlabel("Number of iterations")
    plt.ylabel("Error")
    plt.show()
    # plt.savefig(title+' degree-'+str(i)+'.png')

"""EVALUATION METRICS FUNCTIONS"""

def mean_squared_error(y_hat,y):
  error = (1/len(y)) * np.sum((y_hat - y)**2)
  return error

def mean_absolute_error(y_hat,y):
  error = (1/len(y)) * np.sum(abs(y_hat - y))
  return error

def correlation_coefficient(y_hat,y):
  coef = np.sum((y_hat - y_hat.mean())*(y - y.mean()))/(np.sqrt(np.sum((y_hat-y_hat.mean())**2))*np.sqrt(np.sum((y-y.mean())**2)))
  return coef

"""DATA PRE-PROCESSING"""

from random import Random
data = pd.read_csv("/content/gdrive/MyDrive/FoDS-A1.csv")
Random(14).shuffle(data.values)
X = data.drop("Pressure",axis=1)
y = data["Pressure"]
m = len(X)
X_train = X[:int(m*0.7)]
X_test = X[int(m*0.7):]
y_train = y[:int(m*0.7)]
y_test = y[int(m*0.7):]

"""DRIVER-CODE"""

alpha = [0.001,0.00001]
alpha_stochastic = [0.01,0.000001]
iters = 4000
iters_stochastic = 100
lam = 0.000001
for i in range(10):
    print('\033[1m'+"FOR "+str(i)+" DEGREE POLYNOMIAL"+'\033[0m')
    X_norm = Normalise(X_train)
    X_norm = X_norm.to_numpy()
    X_norm,w = create_poly(X_norm[:,0],X_norm[:,1],i,X_norm)
    X_norm = X_norm[:,2:]
    batch_w, J_his_batch = batch_gradient_descent(X_norm,y_train,w,alpha[int(i/5)],iters)
    #surface_plot(X_norm,y_train,i+1)
    plot('Gradient Descent',iters,J_his_batch,i)
    print("Evaluation Metrics on Training set")
    print_metrics(X_norm,batch_w,y_train)
    print("Evaluation Metrics on Testing set")
    print_metrics(Test(X_test,i),batch_w,y_test)
    batch_w_S, J_his_batch_S = stochastic_gradient_descent(X_norm,y_train,w,alpha_stochastic[int(i/5)],iters)
    #surface_plot(X_norm,y_train,i+1)
    plot('Stochastic Gradient Descent',iters,J_his_batch_S,i)
    print("Evaluation Metrics on Training set")
    print_metrics(X_norm,batch_w_S,y_train)
    print("Evaluation Metrics on Testing set")
    print_metrics(Test(X_test,i),batch_w_S,y_test)
    print("########################################################\n\n")

alpha = 0.00001
alpha_stochastic = 0.000000000000000000000000000000000000000001
for i in np.linspace(0,1,20):
    print('\033[1m'+'Lambda = ' + str(i)+'\033[0m' )
    X_norm = Normalise(X_train)
    X_norm = X_norm.to_numpy()
    X_norm,w = create_poly(X_norm[:,0],X_norm[:,1],9,X_norm)
    X_norm = X_norm[:,2:]
    print("LASSO REGRESSION")
    batch_w, J_his_batch = batch_gradient_descent_L1(X_norm,y_train,w,alpha,4000,i)
    plot('Gradient Descent',4000,J_his_batch,9)
    print_metrics(Test(X_test,9),batch_w,y_test)
    batch_w_S, J_his_batch_S = stochastic_gradient_descent_L1(X_norm,y_train,w,alpha_stochastic,200,i)
    plot('Stochastic Gradient Descent',200,J_his_batch_S,9)
    print_metrics(Test(X_test,9),batch_w_S,y_test)
    print("RIDGE REGRESSION")
    batch_w, J_his_batch = batch_gradient_descent_L2(X_norm,y_train,w,alpha,4000,i)
    plot('Gradient Descent',4000,J_his_batch,9)
    print_metrics(Test(X_test,9),batch_w,y_test)
    batch_w_S, J_his_batch_S = stochastic_gradient_descent_L2(X_norm,y_train,w,alpha_stochastic,200,i)
    plot('Stochastic Gradient Descent',200,J_his_batch_S,9)
    print_metrics(Test(X_test,9),batch_w_S,y_test)
    print("#######################################################\n\n")

#Graph for rms error vs. log(lamba) 
rms_error=[]
lam=[]
for i in np.linspace(0.0000000001,1,20):
  lam.append(i)
  batch_w, J_his_batch = batch_gradient_descent_L2(X_norm,y_train,w,alpha,4000,i)
  rms_error.append(np.sqrt(mean_squared_error(np.dot(Test(X_test,9),batch_w.T),y_test)))
plt.plot(np.log(lam),rms_error)
plt.title('rms error vs log(lambda) Batch L2')
plt.xlabel('log(lambda)')
plt.ylabel('Root Mean Square Error')
plt.show()
print(rms_error)
print(lam)

#Graph for rms error vs. log(lamba) 
rms_error=[]
lam=[]
for i in np.linspace(0.0000000001,1,20):
  lam.append(i)
  batch_w, J_his_batch = batch_gradient_descent_L1(X_norm,y_train,w,alpha,4000,i)
  rms_error.append(np.sqrt(mean_squared_error(np.dot(Test(X_test,9),batch_w.T),y_test)))
plt.plot(np.log(lam),rms_error)
plt.title('rms error vs log(lambda) for Batch L1')
plt.xlabel('log(lambda)')
plt.ylabel('Root Mean Square Error')
plt.show()
print(rms_error)
print(lam)

#Graph for rms error vs. log(lamba) 
rms_error=[]
lam=[]
for i in np.linspace(0.0000000001,1,20):
  lam.append(i)
  batch_w, J_his_batch = stochastic_gradient_descent_L2(X_norm,y_train,w,alpha,200,i)
  rms_error.append(np.sqrt(mean_squared_error(np.dot(Test(X_test,9),batch_w.T),y_test)))
plt.plot(np.log(lam),rms_error)
plt.title('rms error vs log(lambda) for Stochastic L2')
plt.xlabel('log(lambda)')
plt.ylabel('Root Mean Square Error')
plt.show()
print(rms_error)
print(lam)

#Graph for rms error vs. log(lamba) 
rms_error=[]
lam=[]
for i in np.linspace(0.0000000001,1,20):
  lam.append(i)
  batch_w, J_his_batch = stochastic_gradient_descent_L1(X_norm,y_train,w,alpha,200,i)
  rms_error.append(np.sqrt(mean_squared_error(np.dot(Test(X_test,9),batch_w.T),y_test)))
plt.plot(np.log(lam),rms_error)
plt.title('rms error vs log(lambda) for Stochastic L1')
plt.xlabel('log(lambda)')
plt.ylabel('Root Mean Square Error')
plt.show()
print(rms_error)
print(lam)
```
